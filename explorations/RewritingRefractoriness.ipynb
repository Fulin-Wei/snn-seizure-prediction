```python
import matplotlib.pyplot as plt
import math
```

```python
refractorinessDecay = 80
def ref(t):
    return -2 * math.exp(-t/refractorinessDecay)
```

```python
xs = list(range(0, 10))
ys_old = list(map(ref, xs))
ys_new = list(map(lambda t: t/40 - 2, xs))
plt.plot(xs, ys_old, label="Original")
plt.plot(xs, ys_new, label="Approximation")
plt.legend()
plt.show()
```
