```python
import random
import math
randoms = [math.sin(x) + math.cos(2 * x) for x in range(0, 400)]
```

```python
import scipy.signal
```

```python
(f, t, Zxx) = scipy.signal.stft(randoms)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(0, 129), f)
plt.show()
```

```python
def makeplot(plotnum):
    plt.plot(range(129), [abs(Zxx[a, plotnum]) for a in range(129)])
    plt.show()
```
